# Create a Portfolio!

Below we will walk through some basic functionaility available within the ETF Portfolio Manager.


First import the Portfolio API (which gives us access to the ETF and Portfolio classes), and a matplotlib backend for plotting.

```python
import portfolio.api as pf
%matplotlib inline
```

Choose the ETF tickers and the weights you wish to give each ETF in your portfolio.

```python
# Below we give 'CSPX' 50%, with 'IAEA' and 'SUWU' with 25% each
ticker_percent_dict = {'CSPX': 50,
                       'IAEA': 25,
                       'IWDG': 25}

# Create the portfolio object
my_portfolio = pf.Portfolio(ticker_percent_dict, portfolio_name="Cool Portfolio")
my_portfolio
```

### Look at the individual ETFs in your Portfolio

```python
# This will give you each ETF as an ETF object.
etf_list = my_portfolio.etf_list
etf_list
```

```python
# Check out the first few lines of the first ETF
SP500 = etf_list[0]
SP500.df.head(3)
```

```python
# Look at the weightings of individual companies in your ETF:
SP500.get_company_info(company_name="Apple")

# and for your Portfolio
my_portfolio.get_company_info(company_name="Apple")
```

### We can easily see the Sector and Location weightings for each ETF, and the Portfolio as a whole

```python
axes = my_portfolio.plot_all_summarised_ETF(groupby="Sector")

# This can also be done for individual ETFs with: SP500.plot_summarised_ETF(groupby="Sector")
```

```python
# Now lets look at the Portfolio all together
# Use the figsize oarameter to change the size of the figure if you find the text is bunching up.
ax = my_portfolio.plot_summarised_portfolio(groupby="Sector", figsize=(10,8))
```

```python
# Or group by the Location (Geography)
# Use the figsize oarameter to change the size of the figure if you find the text is bunching up.
ax = my_portfolio.plot_summarised_portfolio(groupby="Location",
figsize=(10,10))
```
